```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r'/content/Advertising.csv')
```

```python
print(df.head())
df = df.drop('Unnamed: 0', axis=1)
print(df.head())
```

```python
print("Contain Null values: ",df.isnull().sum())
print("data Info: ",df.info())
print("data Description: ",df.describe())
```

```python
sns.pairplot(df, x_vars=['TV', 'Radio', 'Newspaper'], y_vars='Sales', height=4, aspect=1, kind='scatter')
plt.show()
```

```python
sns.heatmap(df.corr(), annot=True)
plt.show()
```

```python
x = df[['TV','Radio', 'Newspaper']]
y = df['Sales']
print(x.head())
print(y.head())
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
```

```python
print(f"Coefficients: {model.coef_}")
print(f"Intercept: {model.intercept_}")
```

```python
mse = mean_squared_error(y_test, y_pred)
print(f"Mean Squared Error: {mse}")

mae = mean_absolute_error(y_test, y_pred)
print(f"Mean Absolute Error: {mae}")

r2 = r2_score(y_test, y_pred)
print(f"R-squared: {r2}")
print(f"Accuracy: {r2 * 100:.2f}%")
```

```python
plt.scatter(y_test, y_pred)
plt.xlabel("Actual Sales")
plt.ylabel("Predicted Sales")
plt.title("Actual vs Predicted Sales")
plt.show()
```

```python
residuals = y_test - y_pred
sns.histplot(residuals, kde=True)
plt.title("Residuals Distribution")
plt.show()
```

```python
def predict_sales(model):
    print("Enter advertising spends (in thousands of dollars):")
    tv_spend = float(input("TV Advertising Spend: "))
    radio_spend = float(input("Radio Advertising Spend: "))
    newspaper_spend = float(input("Newspaper Advertising Spend: "))
    user_input = pd.DataFrame({
        'TV': [tv_spend],
        'Radio': [radio_spend],
        'Newspaper': [newspaper_spend]
    })

    predicted_sales = model.predict(user_input)[0]
    print(f"\nPredicted Sales Revenue: {predicted_sales:.2f} (in thousands of dollars)")
predict_sales(model)
```
